```python
import pandas as pd
import numpy as np

import requests as rq
from bs4 import BeautifulSoup as bs
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_countries_by_coffee_production'

page = rq.get(url).text
soup = bs(page)
```

```python
table = soup.find('table')
top_10_countries = []
```

```python
rank = []
country = []
sixty_kilogram_bags = []
metric_tons = []
pounds = []
```

```python
for row in table.find_all('tr')[1:]:
    temp = row.text.replace('\n\n', '\t').replace(',','').strip()
    temp_list = temp.split('\t')
    
#     print(temp)
#     print(temp_list)

    rank.append(temp_list[0])
    country.append(temp_list[1])
    sixty_kilogram_bags.append(int(temp_list[2]))
    metric_tons.append(int(temp_list[3]))
    pounds.append(int(temp_list[4]))
```

```python
dict_df = {"Country" : country, 
           "60 Kilogram Bags" : sixty_kilogram_bags, 
           "Metric Tons" : metric_tons, 
           "Pounds" : pounds}
```

```python
df = pd.DataFrame(dict_df, index=rank)
# df.reset_index(drop=True, inplace=True)
# df.set_index(pd.Index(df.Rank))
df
```
