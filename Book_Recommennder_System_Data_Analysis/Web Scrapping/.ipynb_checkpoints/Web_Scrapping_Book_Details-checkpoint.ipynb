```python
def get_google_page_url_with_author(book_name, author_name):
    
#     book_name = book_name[:book_name.find('(') - 1]
    book_name = book_name.replace('&amp;', 'and')
    book_name = book_name.replace(' ', '+')
    
    author_name = author_name.replace(' ', '+')
    
    base_url = "https://www.google.com/search?q="
    url = base_url + book_name + "+" + author_name + "+book"
    
    return url
```

```python
def get_google_page_url(book_name):
    
#     book_name = book_name[:book_name.find('(') - 1]
    book_name = book_name.replace('&amp;', 'and')
    book_name = book_name.replace(' ', '+')
    
    base_url = "https://www.google.com/search?q="
    url = base_url + book_name + "+book"
    
    return url
```

```python
def get_goodread_url(url):
    
    try :
        response = rq.get(url)
        page = response.text
        
        soup = bs(page)
    except:
        print("Sorry goodread url")
        return None
    
    for link in soup.find_all('a'):
        l = link.get('href')
            
        if "/url?q=" in l :
            l = l.replace("/url?q=", "")
        if "goodread" in l :
            if "google" not in l :
                return l[:l.find('&')]
    
    return None
```

```python
def get_synopsis(link):
    
    try :
        response = rq.get(link)
        page = response.text
        
        soup = bs(page)
    except:
        print(link)
        return None
    
    for x in soup.find_all("span", class_='Formatted'):
#     for x in soup.find_all("div", class_='a-expander-content a-expander-partial-collapse-content'):
        return x.text
```

```python
def get_genres(link):
    
    try :
        response = rq.get(link)
        page = response.text
        
        soup = bs(page)
    except:
        print(link)
        return None
    
    for x in soup.find_all("span", class_='BookPageMetadataSection__genre'):
        print(x.text)
```
