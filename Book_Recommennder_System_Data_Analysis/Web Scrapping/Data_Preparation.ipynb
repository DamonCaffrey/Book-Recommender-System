```python
import pandas as pd
import numpy as np

import pickle
```

```python
book_df = pd.read_csv("Books.csv")
rating_df = pd.read_csv("Ratings.csv")
user_df = pd.read_csv("Users.csv")
```

```python
print(book_df.shape)
book_df.head(3)
```

```python
print(rating_df.shape)
rating_df.head(3)
```

```python
print(user_df.shape)
user_df.head(3)
```

```python
book_rating_df = book_df.merge(rating_df, on="ISBN").reset_index(drop=True)

print(book_rating_df.shape)
book_rating_df.head(3)
```

```python
count_users = book_rating_df['User-ID'].value_counts()
count_users
```

```python
count_users[count_users >= 100]
```

```python
desired_user_df = book_rating_df[book_rating_df['User-ID'].isin(count_users[count_users >= 100].index)]

print(desired_user_df.shape)
desired_user_df
```

```python
count_books = desired_user_df['Book-Title'].value_counts()
count_books
```

```python
count_books[count_books >= 50]
```

```python
desired_user_book_df = desired_user_df[desired_user_df['Book-Title'].isin(count_books[count_books >= 50].index)]
desired_user_book_df
```

```python
print(desired_user_book_df['User-ID'].value_counts())
print(desired_user_book_df['Book-Title'].value_counts())
```

```python
count_ratings = count_books.reset_index()

count_ratings.rename(columns={"Book-Title" : "Count-Ratings", "index" : "Book-Title"}, inplace=True)
count_ratings
```

```python
avg_rating = desired_user_book_df.groupby(by='Book-Title').mean()[['Book-Rating']].reset_index().round(2)

avg_rating.rename(columns={"Book-Rating" : "Average-Rating"}, inplace=True)
avg_rating
```

```python
final_rating_df = count_ratings.merge(desired_user_book_df, on='Book-Title')
final_rating_df = avg_rating.merge(final_rating_df, on='Book-Title')
final_rating_df
```

```python
print(desired_user_book_df['User-ID'].value_counts())
print(desired_user_book_df['Book-Title'].value_counts())
```

```python
final_rating_df.to_csv("Final_Rating_df.csv")
```

```python
final_rating_df.groupby(by=['Book-Title', 'ISBN']).mean()
```

```python
final_book_df = final_rating_df.drop_duplicates('Book-Title').reset_index(drop=True)
final_book_df.drop(columns=['User-ID', 'Book-Rating'], inplace=True)

print(final_book_df.shape)
final_book_df.head(3)
```

```python
final_book_df.to_csv("Final_Book_df.csv")
```

```python
ps_df = pd.read_csv("Popular_Synopsis_df.csv")

ps_df.drop(columns=['Unnamed: 0'], inplace=True)
print(ps_df.shape)

ps_df.head(3)
```

```python
print(final_book_df.columns)
print(ps_df.columns)
```

```python
Final_Book_Synopsis_df = final_book_df.merge(ps_df, how='left', on='Book-Title')

print(Final_Book_Synopsis_df.shape)
Final_Book_Synopsis_df.head(3)
```

```python
Final_Book_Synopsis_df.columns
```

```python
Final_Book_Synopsis_df.drop(columns=['Average-Rating_y', 'Count-Ratings_y', 'ISBN_y', 'Book-Author_y', 'Year-Of-Publication_y', 
                                     'Publisher_y', 'Image-URL-S_y', 'Image-URL-M_y', 'Image-URL-L_y',], inplace = True)

Final_Book_Synopsis_df.rename(columns={'Average-Rating_x' : 'Average-Rating', 'Count-Ratings_x' : 'Count-Ratings', 
                                       'ISBN_x' : 'ISBN', 'Book-Author_x' : 'Book-Author', 
                                       'Year-Of-Publication_x' : 'Year-Of-Publication', 'Publisher_x' : 'Publisher', 
                                       'Image-URL-S_x' : 'Image-URL-S', 'Image-URL-M_x' : 'Image-URL-M', 
                                       'Image-URL-L_x' : 'Image-URL-L'}, inplace=True)

print(Final_Book_Synopsis_df.shape)
Final_Book_Synopsis_df.head(3)
```
